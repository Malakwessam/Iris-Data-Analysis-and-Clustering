# Clustering

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
```

```python
X = df.drop(columns='species') # to keep only numerical columns
```

```python
X.head()
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_scaled = pd.DataFrame(X_scaled, columns=X.columns)#learn and apply scaling
```

```python
X_scaled.head()
```

# Elbow method

```python
wcss = []

for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X_scaled)
    wcss.append(kmeans.inertia_)
```


*  KMeans() → create model
*  n_clusters → number of clusters

*  fit() → train model
*   inertia_ → evaluate clustering

A loop was used to apply K‑Means clustering for different numbers of clusters (K = 1 to 10). For each value of K, the model was fitted to the standardized data, and the inertia value was recorded. These values were used to construct the elbow plot in order to determine the optimal number of clusters.

```python

plt.figure(figsize=(6, 4))
plt.plot(range(1, 11), wcss, marker='o' ,color = 'pink')
plt.title('Elbow Method for Optimal K')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('Inertia')
plt.show
```

```python

kmeans = KMeans(n_clusters=3, random_state=42)
clusters = kmeans.fit_predict(X_scaled)

df['Cluster'] = clusters
```

```python
plt.figure(figsize=(6, 5))

sns.scatterplot(
    data=df,
    x='petal_length',
    y='petal_width',
    hue='Cluster',
)

plt.title('K-Means Clustering (Petal Features)')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Petal Width (cm)')
plt.legend(title='Cluster')

plt.show()
```

```python
pd.crosstab(df['species'], df['Cluster'])
```

```python
df.head()
```

The clustering results differ from the actual species labels because K‑Means groups data based on feature similarity rather than predefined labels. In this case, Versicolor and Virginica were grouped together due to their overlapping characteristics, while Setosa was split into two clusters. This indicates that the data does not form perfectly separable clusters and highlights limitations of K‑Means when clusters are not clearly distinct.
